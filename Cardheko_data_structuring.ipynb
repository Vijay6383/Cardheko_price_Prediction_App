```python
# import necessary modules
import pandas as pd
import numpy as np
import ast
```

Load Data 

```python
df = pd.read_excel('bangalore_cars.xlsx')
df.head()
```

```python
# car_link is unrelated to price so drop it
df.drop('car_links', axis=1, inplace=True)
```

```python
df.head()
```

```python
# convert all string to dict
def convert_to_dict(cell):
       return ast.literal_eval(cell)

for col in df.columns:
    df[col] = df[col].apply(convert_to_dict)
```

## New Car Detail

```python
# json_normalize will convert unstructured dict to table structure
df_car_details = pd.json_normalize(df['new_car_detail'])
df_car_details.head()
```

## New Car Overview

```python
df_car_overview = pd.json_normalize(df['new_car_overview'])
df_car_overview.head()
```

```python
# function to get key and value in seperate dict
def convert_top_details(details):
    return {item['key']: item['value'] for item in details}
```

```python
# apply function to all values in top column
df_top = df_car_overview['top'].apply(convert_top_details).apply(pd.Series)
df_top.head()
```

```python
df_Car = pd.concat([df_car_details, df_top], axis=1)
df_Car.head()
```

## New car feature

```python
df_car_feature = pd.json_normalize(df['new_car_feature'])
df_car_feature.head()
```

```python
# function to count list of top features
def count_top_features(features):
    if isinstance(features, list):
        return len(features)
    else:
        return 0
```

```python
df_car_feature['top'] = df_car_feature['top'].apply(count_top_features)
df_car_feature.head()
```

```python
df_car_feature_data = pd.json_normalize(df_car_feature['data'])

df_car_feature_data.head()
```

```python
df_car_feature_data['comfort_count'] = pd.json_normalize(df_car_feature_data[0])['list']
df_car_feature_data['comfort_count'] = df_car_feature_data['comfort_count'].apply(count_top_features)
df_car_feature_data['comfort_count'].head()
```

```python
df_car_feature_data['Safety_count'] = pd.json_normalize(df_car_feature_data[3])['list']
df_car_feature_data['Safety_count'] = df_car_feature_data['Safety_count'].apply(count_top_features)
df_car_feature_data['Safety_count'].head()
```

```python
df_car_feature_data['Entertainment_count'] = pd.json_normalize(df_car_feature_data[4])['list']
df_car_feature_data['Entertainment_count'] = df_car_feature_data['Entertainment_count'].apply(count_top_features)
df_car_feature_data['Entertainment_count'].head()
```

```python
df_car_feature_data.drop([0, 1, 2, 3, 4], axis=1, inplace=True)
df_car_feature_data.head()
```

```python
df_car_feature_data = pd.concat([df_car_feature_data, df_car_feature['top']], axis=1)
df_car_feature_data.rename(columns={'top': 'top_features_count'}, inplace=True)
df_car_feature_data.head()
```

## New Car Specifications

```python
df_car_specs = pd.json_normalize(df['new_car_specs'])
df_car_specs.head()
```

```python
df_car_specs_data = pd.json_normalize(df_car_specs['data'])
df_car_specs_data.head()
```

```python
df_car_specs_data1 = pd.json_normalize(df_car_specs_data[0])['list'].apply(convert_top_details).apply(pd.Series)
df_car_specs_data1.head()
```

```python
df_banglore_cars = pd.concat([df_Car, df_car_feature_data, df_car_specs_data1], axis=1)
df_banglore_cars.head()
```

```python
df_banglore_cars['City'] = 'Bangalore'
df_banglore_cars.head()
```
